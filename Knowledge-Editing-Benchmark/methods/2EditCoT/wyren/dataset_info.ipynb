## 查看数据集内容

1. 查看数据集元信息（不下载数据）

```python
from huggingface_hub import dataset_info
import json

# 获取数据集信息
info = dataset_info("zjunlp/KnowEdit")

# 查看README
print("README内容:")
print(info.card_data)  # 包含使用说明

# 查看数据集配置
print("\n数据集配置:")
for config in info.card_data.get("configs", []):
    print(f"  - {config}")
```

2. 下载小样本查看结构

```python
from datasets import load_dataset

# 方法1：用streaming模式查看第一条数据
dataset = load_dataset("zjunlp/KnowEdit", streaming=True)
for example in dataset["train"].take(1):
    print("第一条数据的结构:")
    print(example)
    print("\n字段名:", list(example.keys()))
```
